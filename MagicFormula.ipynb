Vamos fazer um ranking das ações utilizando a Magic Formula de Joel Greenblatt utilizando Python

Fazendo os imports

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import yfinance as yf
import string
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'http://www.fundamentus.com.br/resultado.php'
agent = {"User-Agent": "Mozilla/5.0"}
r = requests.get(url, headers = agent)
```

```python
df = pd.read_html(r.text,  decimal=',', thousands='.')[0]
```

Alterando valores com %

```python
for coluna in ['Div.Yield', 'Mrg Ebit', 'Mrg. Líq.', 'ROIC', 'ROE', 'Cresc. Rec.5a']:
  df[coluna] = df[coluna].str.replace('.', '')
  df[coluna] = df[coluna].str.replace(',', '.')
  df[coluna] = df[coluna].str.rstrip('%').astype('float') / 100
```

```python
df
```

Agora, vamos filtrar os ativos que tenham, no mínimo, 1 milhão de volume diário negociado

```python
maiorLiquidez = df[df['Liq.2meses'] >= 1000000]
maiorLiquidez
```

Excluindo empresas com Patrimônio Líquido negativo, EBIT negativo e ROIC negativo

```python
patrimonioLiquido = maiorLiquidez[maiorLiquidez['Patrim. Líq'] > 0]
patrimonioLiquido = maiorLiquidez[maiorLiquidez['EV/EBIT'] > 0]
patrimonioLiquido = maiorLiquidez[maiorLiquidez['ROIC'] > 0]
patrimonioLiquido
```

```python
dataNovo = patrimonioLiquido[['Papel', 'ROIC', 'EV/EBIT']]
dataNovo
```

Vamos ordenar por Earning Yield

```python
EYOrdenado = dataNovo.sort_values(by = 'EV/EBIT')
EYOrdenado = EYOrdenado.reset_index(drop = True)
EYOrdenado['EYOrdenado'] = EYOrdenado.index
EYOrdenado
```

Agora vamos ordernar por ROIC

```python
ROICOrdenado = EYOrdenado.sort_values(by = 'ROIC', ascending= False)
ROICOrdenado = ROICOrdenado.reset_index(drop = True)
ROICOrdenado['ROICOrdenado'] = ROICOrdenado.index
dados = ROICOrdenado
dados
```

Criando um Score para rankear as melhores empresas

```python
dados['Score'] = dados['ROICOrdenado'] + dados['EYOrdenado']
dados.sort_values(by = 'Score')
```

Utilizando o Yahoo Finance para coletar o setor do ativo

```python
def getSetor(ativo):
  setor=yf.Ticker(ativo)
  return setor.info['sector']
```

```python
symbols = dados['Papel'] + '.SA'

sectorDict = {}

for tick in symbols:
  try:
    print("Coletando setor do ativo: ", tick)
    sectorDict[tick] = getSetor(tick)
  except:
    print("Erro ao coletar setor de: ", tick)
    sectorDict[tick] = 'NA'

stockSectorDF = pd.DataFrame.from_dict(sectorDict, orient = 'index').reset_index()
stockSectorDF.columns = ['Papel', 'sector']

stockSectorDF
```

Filtrando setores indesejados para a Magic Formula

```python
dados['Papel'] = dados['Papel'] + '.SA'
dadosNovos = pd.merge(dados, stockSectorDF, on = 'Papel', how = 'inner')
filtroSetores = dadosNovos[dadosNovos['sector'] != 'Financial Services']
filtroSetores = filtroSetores[filtroSetores['sector'] != 'Energy']
filtroSetores = filtroSetores[filtroSetores['sector'] != 'Utilities']
filtroSetores = filtroSetores[filtroSetores['sector'] != 'NA']
filtroSetores
```

Gerando o rank definitivo ordenado da Magic Formula com as 30 melhores ações
